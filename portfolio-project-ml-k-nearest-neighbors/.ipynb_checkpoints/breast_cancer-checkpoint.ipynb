```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import matplotlib.pyplot as plt
```

## Load and investigate the data

```python
breast_cancer_data = load_breast_cancer()
```

```python
print(breast_cancer_data.data[0])
```

```python
print(breast_cancer_data.feature_names)
```

```python
print(breast_cancer_data.target)
```

```python
print(breast_cancer_data.target_names)
```

## K nearest neighbors

```python
training_data, validation_data, training_labels, validation_labels = train_test_split(breast_cancer_data.data, breast_cancer_data.target, train_size = 0.2, random_state = 100)

classifier = KNeighborsClassifier(n_neighbors = 3)
classifier.fit(training_data, training_labels)

print(classifier.score(validation_data, validation_labels))

accuracies = []
for i in range(1, 101):
  classifier = KNeighborsClassifier(n_neighbors = i)
  classifier.fit(training_data, training_labels)
  accuracies.append(classifier.score(validation_data, validation_labels))#15

k_list = range(1, 101)

plt.plot(k_list, accuracies)
plt.xlabel('k')
plt.ylabel('Validation Accuracy')
plt.title('Breast Cancer Classifier Accuracy')
plt.show()
plt.clf()
```
