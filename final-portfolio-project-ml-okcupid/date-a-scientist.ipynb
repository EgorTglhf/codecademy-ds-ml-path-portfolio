# Preprocess data

Time to select features we will use to predict labels. Such features as age, height, income obviously are not good indicators for such labels as the zodiac sign. We are considering the probable connection between zodiac signs and habits and tendencies of human character.

```python
columns = ['body_type', 'diet', 'drinks', 'drugs', 'education', 'job', 'offspring', 'orientation', 'pets', 'religion_name', 'smokes', 'zodiac']

data = profiles[columns]
```

## Missing data

```python
data.isnull().sum()
```

There is a lot of missing data. The entire volume of data is large enough to just drop missing values.

```python
data = data.dropna()
```

## Dummy values

One of good encoding technique is a One-hot Encoding which is great for categorical variables.

```python
for column in columns[:-1]:
    data = pd.get_dummies(data, columns=[column], prefix=[column])
```

## Label imbalance

The distribution of the prediction labels needs to be examined for imbalance, as this is crucial in a multi-class problem where multiple outcomes are possible. An imbalanced response variable is problematic because certain labels may appear infrequently, making it difficult for machine learning algorithms to learn effectively, leading to poor predictions.

```python
data.zodiac.value_counts()
```

In this dataset, the counts of all zodiac signs are relatively even, with no significant deviations. Therefore, we do not need to be concerned about label imbalance or taking steps to address it.

## Splitting data

So we are almost ready to implement training and validation of models. The last one we should do before is to split data into training and validation (test) data.

```python
data.head()
```
